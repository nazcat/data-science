```python
import pandas as pd

df = pd.read_csv(r'C:\Users\IBM_ADMIN\Documents\Spotify\playlist_summary_external.csv', encoding='utf-8')
df.head(10)
```

```python
df = df.join(pd.DataFrame(
    {
        'd_skip_rate': df['skippers']/df['dau']
    }, index=df.index
))

df = df.join(pd.DataFrame(
    {
        'm_stream_rate': df['mau']/df['users']
    }, index=df.index
))

df = df.join(pd.DataFrame(
    {
        'stream_30s_rate': df['stream30s']/df['streams']
    }, index=df.index
))

df = df.join(pd.DataFrame(
    {
        'track_update_rate': df['n_local_tracks']/df['n_tracks']
    }, index=df.index
))

df = df.join(pd.DataFrame(
    {
        'd_30s_stream_per_user': df['stream30s']/df['dau']
    }, index=df.index
))

df = df.join(pd.DataFrame(
    {
        'd_tot_stream_per_user': df['streams']/df['dau']
    }, index=df.index
))

df = df.join(pd.DataFrame(
    {
        'd_tot_stream_per_user': df['streams']/df['dau']
    }, index=df.index
))

df.head(10)
```

```python
import pandasql as ps

df_g = """
    SELECT 
        (genre_1||'_'||genre_2) AS genre_combo
        ,dau
        ,mau
        ,streams
        ,stream30s
        ,monthly_stream30s
        ,users
        ,n_tracks
        ,n_artists
        ,n_albums
        
        FROM df
        
        UNION ALL
        
        SELECT 
        (genre_2||'_'||genre_3) AS genre_combo
        ,dau
        ,mau
        ,streams
        ,stream30s
        ,monthly_stream30s
        ,users
        ,n_tracks
        ,n_artists
        ,n_albums
        
        FROM df
        
        UNION ALL
        
        SELECT 
        (genre_1||'_'||genre_3) AS genre_combo
        ,dau
        ,mau
        ,streams
        ,stream30s
        ,monthly_stream30s
        ,users
        ,n_tracks
        ,n_artists
        ,n_albums
        
        FROM df; """

df_genre = (ps.sqldf(df_g, locals()))
df_genre.head(5)
```

```python
df_genre.to_csv('C:\\Users\\IBM_ADMIN\\Documents\\Spotify\\genre_combo_streams.csv', header=True, index=False, index_label=None)
```
